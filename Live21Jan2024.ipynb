```python
X=df[['num_loves','num_reactions']]
X.head()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
  k_means=KMeans(n_clusters=i,init='k-means++')
  k_means.fit(X)
  wcss.append(k_means.inertia_)

plt.plot(range(1,11),wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('k values')
plt.ylabel('wcss')
```

```python
# Create k means model alogorithms
#a) using cluster 6
k_means1=KMeans(n_clusters=6, init='k-means++', random_state = 42)
k_means1.fit(X)

```

```python
#Add clustering to the dataframe
df['Segmented kmeans1']= k_means1.labels_
df.head()
```

```python
#Grouping the Cluster
df_segmented_group= df.groupby(['Segmented kmeans1']).mean()

df_segmented_group
```

```python
#Add segment to the labels
df['Labels'] = df['Segmented kmeans1'].map({0: 'cluster 1',
                                            1:'cluster 2',
                                            2:'cluster 3',
                                            3:'cluster 4',
                                            4:'cluster 5',
                                            5:'cluster 6'})

df.head(20)
```

```python
x_axis=df['num_loves']
y_axis=df['num_reactions']
plt.figure(figsize=(10,8))
sns.scatterplot(x= x_axis,y= y_axis, hue= df['Labels'], data=df)
```

```python
# Create k means model alogorithms
#b) using cluster =5
k_means2=KMeans(n_clusters=5, init='k-means++', random_state = 42)
k_means2.fit(X)

#Add clustering to the dataframe
df['Segmented kmeans2']= k_means2.labels_
df.head()

#Add segment to the labels
df['Label'] = df['Segmented kmeans2'].map({0: 'cluster 1',
                                            1:'cluster 2',
                                            2:'cluster 3',
                                            3:'cluster 4',
                                            4:'cluster 5',
                                           })


x_axis=df['num_loves']
y_axis=df['num_reactions']
plt.figure(figsize=(10,8))
sns.scatterplot(x= x_axis,y= y_axis, hue= df['Label'], data=df)
```
