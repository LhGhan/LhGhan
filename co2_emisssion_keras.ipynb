```python

import pandas as pd

df=pd.read_csv('FuelConsumption.csv',na_values="?")

df.head()
```

```python
df.isnull().sum()
```

```python
df['FUELTYPE'].value_counts()
```

```python
df.replace('X',1,inplace=True)
df.replace('Z',2,inplace=True)
df.replace('E',3,inplace=True)
df.replace('D',4,inplace=True)
```

```python
df.columns
```

```python
concrete_df_columns= df[['ENGINESIZE', 'CYLINDERS',
      'FUELCONSUMPTION_CITY', 'FUELTYPE',
       'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB',
      'CO2EMISSIONS']]

concrete_df_columns.head()
```

```python
predictors= df[['ENGINESIZE', 'CYLINDERS',
      'FUELCONSUMPTION_CITY', 'FUELTYPE',
       'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB']]

predictors.head()
```

```python
target=df['CO2EMISSIONS']
target.head()
```

```python
#2. Data Normalization
#normalization
#using z-score

predictors_norm= (predictors-predictors.mean())/predictors.std()

predictors_norm.head()
```

```python
#number of features / predictors
n_cols = predictors_norm.shape[1]
n_cols
```

```python
#3.Modeling using Keras
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#let define a function for regression, showing number of neurone

def regression_model():
  #creating the model

  model=Sequential()

  #creating the input layer
  model.add(Dense(50,activation='relu',input_shape=(n_cols,)))

  #creating the hidden layer
  model.add(Dense(50,activation='relu'))
  model.add(Dense(50,activation='relu'))


  #create output layer
  model.add(Dense(1))

  #compile the model
  model.compile(optimizer='adam',loss='mean_squared_error')
  return model
```

```python
#4.Train and test deep learning model
model = regression_model()

model.fit(predictors_norm,target,validation_split=0.3, epochs=40,verbose=2)
```
